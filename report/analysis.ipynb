## Setup

```python
import matplotlib.pyplot as plt
import geopandas as gpd
from pathlib import Path
```

## Import

```python
sample_path = Path('../data/vias_estudo.geojson')
vias_sample = gpd.read_file(sample_path)
```

## AnÃ¡lise

```python
mask = vias_sample['HIERARQUIA'].isin(['2', '3', '4'])
vias_sample = vias_sample[mask]
trechos_acima_vel = vias_sample.query('spd_kmh > 60')
```

```python
vias_com_radar = vias_sample[vias_sample['presenca_radar'] == True]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10), dpi=300)

vias_sample.plot(ax=ax1, color="#9a9a9a", linewidth=0.5)
trechos_acima_vel.plot(ax=ax1, color="red", linewidth=0.8)

vias_sample.plot(ax=ax2, color="#9a9a9a", linewidth=0.5)
vias_com_radar.plot(ax=ax2, color="blue", linewidth=0.8)
```
